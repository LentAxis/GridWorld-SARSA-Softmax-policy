## Classic Tabular Reinforcement Learning

### Gridworld

This Notebook contains a simple implementation of the SARSA Algorithm with a Softmax Policy. The Environment and the Agent are customizable. I will do my best to explain what I did here. 

The code uses Numpy and Matplotlib

---

If you don't change anything, you will get an environment, that looks like the picture below. The blue diamond represents the startingpoint. The green one the Goal. The grey sections represent walls and act as termination conditions. That means, if our Agent steps on a wall, he will get a penalty and has to start again.

<img src="pictures/Gridworld_V2.png" width="400" align="left"/>




Let's get into the interesting part.

### Environment

First, we will create the environment. For that, we will need to initiate some parameters.

```python
# Tuple with the number of cells in x and y
size = (12, 7) 

# Starting cell (uses the index, so it starts with 0)
start=(0, 0)

# Goal cell (uses the index, too)
goal=(8, 5) 

# Penalty for every step taken
step_penalty=-0.25

# Reward if the Agent reaches the goal
goal_reward=50
```

Now we can initiate the Environment class. We get the class from the *GridWorld.py* file. If you saved the file in an other folder, you will need to change the import line.

```python
import GridWorld as gw
Gridworld_Environment = gw.Environment(size=size, start=start, goal=goal, step_penalty=step_penalty, goal_reward=goal_reward)
```

With the *add_obstacle((x, y), penalty)* method we can add walls and assign them a penalty value.

```python
# For x = 1
Gridworld_Environment.add_obstacle((1, 0), -10)
Gridworld_Environment.add_obstacle((1, 2), -10)
Gridworld_Environment.add_obstacle((1, 3), -10)
Gridworld_Environment.add_obstacle((1, 5), -10)

# For x = 2
Gridworld_Environment.add_obstacle((2, 5), -10)

# For x = 3
Gridworld_Environment.add_obstacle((3, 2), -10)
Gridworld_Environment.add_obstacle((3, 3), -10)
Gridworld_Environment.add_obstacle((3, 4), -10)
Gridworld_Environment.add_obstacle((3, 5), -10)

# For x = 4
Gridworld_Environment.add_obstacle((4, 1), -10)
Gridworld_Environment.add_obstacle((4, 2), -10)

# For x = 5
Gridworld_Environment.add_obstacle((5, 2), -10)
Gridworld_Environment.add_obstacle((5, 4), -10)
Gridworld_Environment.add_obstacle((5, 5), -10)
Gridworld_Environment.add_obstacle((5, 6), -10)

# For x = 6
Gridworld_Environment.add_obstacle((6, 2), -10)

# For x = 7
Gridworld_Environment.add_obstacle((7, 2), -10)
Gridworld_Environment.add_obstacle((7, 4), -10)
Gridworld_Environment.add_obstacle((7, 5), -10)

# For x = 8
Gridworld_Environment.add_obstacle((8, 4), -10)

# For x = 9
Gridworld_Environment.add_obstacle((9, 3), -10)
Gridworld_Environment.add_obstacle((9, 4), -10)
Gridworld_Environment.add_obstacle((9, 5), -10)

# For x = 10
Gridworld_Environment.add_obstacle((10, 5), -10)

# For x = 11
Gridworld_Environment.add_obstacle((11, 2), -10)
```

Et voil√†! The Environment is done and ready for the Agent.

### Agent

First, we do the same as we did for the environment.

```python
# Learning Rate
alpha=0.1

# Discount value
gamma=0.25

# Information output while running
printinfo=False

# Number of episodes
episodes=10000

# Number of episodes without tau decay with gamma = 1.0
# These settings will be true for the last episodes
constant_episodes=2000

# Number of steps before termination
steps=100

# Temperature parameter at the beginning
tau=100.0

# Temperature parameter at it's lowest
tau_min=1.0
```

And we can initialise the Agent and let him run.

```python
Gridworld_Agent = gw.Agent(Gridworld_Environment, alpha=alpha, gamma=gamma, printinfo=printinfo)
```

```python
Gridworld_Agent.run(episodes=episodes, constant_episodes=constant_episodes, steps=steps, tau=tau, tau_min=tau_min)
```

To check, if the tau value decay over time fits your expectations, we will plot it out real fast.

```python
Gridworld_Agent.plot_tau_decay()
```

To see what the Agent has learned from this, we can plot the environment with the Q-Values for each action at each cell. 

Try not to puke when you see the color choice for the plot. 

```python
Gridworld_Agent.plot_grid()
```
